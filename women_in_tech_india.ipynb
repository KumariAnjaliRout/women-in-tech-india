```python
import pandas as pd

df = pd.read_csv('women_in_tech_full.csv')


df.head()
```

```python
import matplotlib.pyplot as plt

df_sorted = df.sort_values(by='Female_Percentage', ascending=False)

plt.figure(figsize=(14, 8))
plt.barh(df_sorted['State'], df_sorted['Female_Percentage'], color='orchid')
plt.xlabel('Female Participation (%)')
plt.title('Women in Tech - Percentage of Female Students by State')
plt.gca().invert_yaxis()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python

top_states = df.sort_values(by='Female_Students', ascending=False).head(10)

plt.figure(figsize=(8, 8))
plt.pie(top_states['Female_Students'], labels=top_states['State'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 States by Number of Female Students in Tech')
plt.axis('equal')
plt.show()
```

```python
df = pd.read_csv("women_in_tech_full.csv")
df = df.rename(columns={"State": "id", "Female_Percentage": "value"})
```

```python
import pandas as pd

# Define the data
data = {
    "State": [
        "Andhra Pradesh", "Arunachal Pradesh", "Assam", "Bihar", "Chhattisgarh", "Goa",
        "Gujarat", "Haryana", "Himachal Pradesh", "Jharkhand", "Karnataka", "Kerala",
        "Madhya Pradesh", "Maharashtra", "Manipur", "Meghalaya", "Mizoram", "Nagaland",
        "Odisha", "Punjab", "Rajasthan", "Sikkim", "Tamil Nadu", "Telangana", "Tripura",
        "Uttar Pradesh", "Uttarakhand", "West Bengal", "Delhi"
    ],
    "Total_Students": [
        74000, 8000, 35000, 69000, 27000, 9000, 71000, 45000, 8000, 27000, 88000, 62000,
        67000, 100000, 5000, 4000, 3000, 2500, 63000, 42000, 80000, 2000, 95000, 72000,
        6000, 91000, 32000, 77000, 55000
    ],
    "Female_Students": [
        31000, 3100, 14500, 19000, 11500, 4200, 22000, 16000, 3200, 10000, 32000, 29000,
        23000, 37000, 2400, 1600, 1200, 1100, 26000, 18000, 27000, 900, 41000, 25000,
        2600, 28000, 13000, 34000, 23000
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate Female Percentage
df["Female_Percentage"] = (df["Female_Students"] / df["Total_Students"]) * 100
```

```python
import plotly.express as px

# Sort and plot
fig = px.bar(
    df.sort_values(by="Female_Percentage", ascending=False),
    x="Female_Percentage",
    y="State",
    orientation="h",
    color="Female_Percentage",
    color_continuous_scale="reds",
    title="Female Participation (%) in Tech by State"
)

fig.update_layout(
    xaxis_title="Female Participation (%)",
    yaxis_title="State",
    yaxis=dict(categoryorder='total ascending')
)

fig.show()
```

```python

df = pd.DataFrame(data)

# Sort and get bottom 10 states
bottom_10 = df.sort_values(by="Female_Students", ascending=True).head(10)

# Plot bar chart
fig = px.bar(
    bottom_10,
    x="Female_Students",
    y="State",
    orientation="h",
    title="Bottom 10 States by Number of Female Students in Tech",
    color="Female_Students",
    color_continuous_scale="reds"
)

fig.update_layout(
    xaxis_title="Number of Female Students",
    yaxis_title="State",
    yaxis=dict(categoryorder="total ascending")
)

fig.show()
```

```python
df = pd.DataFrame(data)

# Sort and get top 10 states
top_10 = df.sort_values(by="Female_Students", ascending=False).head(10)

# Bar chart
fig = px.bar(
    top_10,
    x="Female_Students",
    y="State",
    orientation="h",
    title="Top 10 States by Number of Female Students in Tech",
    color="Female_Students",
    color_continuous_scale="purples"
)

fig.update_layout(
    xaxis_title="Number of Female Students",
    yaxis_title="State",
    yaxis=dict(categoryorder="total ascending")
)

fig.show()
```
